# Git in Other Environments - Graphical Interfaces

Gits native environment is in the terminal. New features show up there first, and only at the command line is the full power of Git completely at your disposal. But plain text isnt the best choice for all tasks; sometimes a visual representation is what you need, and some users are much more comfortable with a point-and-click interface.

```sudo apt install gitk git-gui```
