## Creating Tags

Git supports two types of tags:

1. Lightweight Tag
2. Annotated Tags

#### 1. Lightweight Tag

This is basically the commit checksum stored in a fileno other information is kept.





#### 2. Annotated Tag

Creating an annotated tag in Git is simple. The easiest way is to specify -a when you run the tag command:

This is Acutally a Better Way, because it allow to provide information about each tag like tag name & tag messages.



## Updating Tag in Remote Github

So, we updated a tag in git cli now its time to update that tag to github, so lets do it !

```git push <remote-name> <tag-no>```
```git push <remote-name> --tags```

## Deleting the Tag

To delete a tag on your local repository, you can use,

```git tag -d <tagname>```



```git push <remote-name> -d <tagname>```





## Renaming the Tag


To rename a tag on your local repository, you can use,

```git tag <newname> <oldname>```

*delete tag after that*
