#### Merge Conflicts

The most important part in the phrase of merging ***merge conflicts***
this process doesnt go smoothly. If you changed the same part of the same file differently in the two branches youre merging, Git wont be able to merge them cleanly. 

Lets Take an Example,

1. in git_commands branch we have a file SampleChanged now we changed the file content, 



2. we commit the changes and switch to testing



3. We Again Change the Same File Content



4. Now we have to switch to git_commands branch and start merging the testing branch to git_commands branch.



5. This is what we called merge confilcts

***when we changed the content of files at same place of two merging branches***

#### Merge Conflict Resolve

1. We can use ide like vscode to resolve merge conflict fast and easy way, just select current changes or incoming changes.



2. Or we can just update files by ourselfes for conflict update

#### Deleting a Branch

For Deleting a Branch we have a command

```git branch -d <branchname>```
```git branch -D <branchname>```

do not confuse with -d & -D, we use -d when the merge is complete and the branches are at same HEAD, but when you don't want to merge and delete with out Merge -D means force delete is used.
